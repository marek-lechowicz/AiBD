```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2+5

x1 = np.linspace(-1, 1, 100)
x2 = np.linspace(-6, 6, 100)
x3 = np.linspace(0, 5, 100)

plt.figure()
plt.plot(x1, f(x1))
plt.title("x>-1 oraz x<1")
plt.show()

plt.figure()
plt.plot(x2, f(x2))
plt.title("x>-6 oraz x<6")
plt.show()

plt.figure()
plt.plot(x3, f(x3))
plt.title("x>0 oraz x<5")
plt.show()
```
