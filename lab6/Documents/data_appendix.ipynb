# Data appendix

## 1. Zmienne w pliku AnalysisData/wielkopolskie.csv zawierają dane na temat odkurzaczy oraz informacje o kupujących

```python
import pandas as pd
import matplotlib.pyplot as plt

w = pd.read_csv('../AnalysisData/wielkopolskie.csv', index_col=0)
w.info()
```

### 1.1 Dni od zakupu


Opis: liczba dni, które upłynęły od zakupu odkurzacza

Liczba wartości non-null: 534

```python
plt.figure()
w['Dni od zakupu'].hist(bins=15)
plt.title('Histogram dla zmiennej "Dni od zakupu"')
plt.show()

print("Podstawowe dane statystyczne")
print(w['Dni od zakupu'].describe())
```

### 1.2 Marka 


Opis: znak firmowy umieszczony na produkcie

Liczba wartości non-null: 534

```python
plt.figure()
w['Marka'].value_counts().plot(kind='bar')
plt.title('Wykres słupkowy dla zmiennej Marka')
plt.show()

print("Frequency table\nValue      freq")
print(w['Marka'].value_counts())
```

### 1.3 Wiek kupującego


Opis: Wiek kupującego wyrażonych w latach - pomimo tego, że jest wyrażony jako float64 nie uwzględnia się niepełnych lat.

Liczba wartości non-null: 473

```python
plt.figure()
w['Wiek kupującego'].hist(bins=35)
plt.title('Histogram dla zmiennej "Wiek kupującego"')
plt.show()

print("Podstawowe dane statystyczne")
print(w['Wiek kupującego'].describe())
```

### 1.4 Płeć kupującego

Opis: Płeć kupującego wyrażona literą K - kobieta, M - mężczyzna

Liczba wartości non-null: 519

```python
plt.figure()
w['Płeć kupującego'].value_counts().plot(kind='bar')
plt.title('Wykres słupkowy dla zmiennej Płeć kupującego')
plt.show()

print("Frequency table\nValue      freq")
print(w['Płeć kupującego'].value_counts())
```

### 1.5 Ocena

Opis: Ocena odkurzacza przez kupującego. Dopuszczalne są wartości od 0.0 do 5.0, co pół punktu.

Liczba wartości non-null: 534

```python
plt.figure()
w['Ocena'].hist(bins=11)
plt.title('Histogram dla zmiennej "Ocena"')
plt.show()

print("Podstawowe dane statystyczne")
print(w['Ocena'].describe())
```
